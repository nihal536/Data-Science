```python
import numpy as np  # Import Numpy Library for numerical operation (arrays, math functions)
```

```python
sample=pd.read_csv("Bulbs_Lifespan.csv")  # Read the CSV file and store it as a DataFrame named 'sample'
```

```python
sample            # Display the dataset (table) to view its contents
```

```python
#----------------------------------------------------------------------------------------------------------------
```

```python
import pandas as pd  # Import Pandas for handling data (DataFrame)
```

```python
from scipy.stats import ttest_1samp  # Import function to perform one-sample t-test
                                     # By default, this perform a TWO-TAILED TEST (checks mean != uo )
```

```python
# Hypothesized population mean
pop_mean=1000   # Define uo (null hypothesis mean = 1000)
                # Ho: u = 1000
                # H1: u != 1000 -> TWO-TAILED TEST
```

```python
# Perfrom the one-sample t-test
t_statistic,p_value=ttest_1samp(sample,pop_mean)  # Apply t-test using formula:
                                                  # t_statistics -> measure how far sample mean is from hypothesized mean
                                                  # p_value -> probability of observing this result if Ho is true
```

```python
# significance level
alpha=0.05       # Set a = 0.05 (5% risk of Type I error)
                 # For TWO-TAILED TEST -> α/2 = 0.025 in each tail
```

```python
# RESULTS
print("T-Statistics: ", np.round(t_statistic,4)) #Print the t-statistics rounded to 4 decimal places (shows deviation from hypothesized mean)

print("p-value: ",np.round(p_value,4)) # Print the p-value rounded to 4 decimal places (probability under Ho)

if p_value < alpha:                    # Check if p-value is less than significance level (α = 0.05)
    print("Reject the null hypothesis: The mean lifespan is significantly different from 1000 hours")
                                       # If True -> strong evidence against Ho -> significant different exists
    
else:
    print("Fail to reject the null hypothesis: No significant difference from 1000 hours")
                                       # If False -> not enough evidence -> no significant difference
```

```python
#For alternate hypothesis: Less than (H1:H < 1000)
#This means we are performing a LEFT-TAILED t-test.
```

```python
if t_statistic < 0:   # Check if the statistics is negative -> supports u < 1000 direction
                      # (If it's positive, it contradicts the alternative hypothesis)
    
    one_tailed_p_value = p_value/2   # Convert two-tailed p-value to one-tailed
                                     # because we only care about the left side

    if one_tailed_p_value<alpha:     # Compare p-value with significance level (e.g., 0.05)
                                     # If smaller -> strong evidence against null hypothesis
        
        print("Reject the null hypothesis: The mean lifespan is significantly less than 1000 hours.")
                                     # Conclusion: Data provides enough evidence that u < 1000
        
    else:
        print("Fail to reject the null hypothesis: No significant evidence that the mean lifespan is less than 1000 hours.")
                                     # Conclusion: Evidence is not strong enough to support u < 1000
else:
    print("Fail to reject the null hypothesis: Test statistics does not support H1(mean<1000)")
                                     # If t_statistic is > 0 -> result goes in the opposite direction
                                     # So we do NOT even consider rejecting H0, regardless of p-value
```

```python
# ONE SAMPLE Z TEST
```

```python
import numpy as np   # Import NumPy for numerical operations (arrays, math function)
```

```python
import pandas as pd  # Import Pandas for handling datasets (Like CSV files)
```

```python
df = pd.read_csv("Bulbs_Lifespan.csv")  # Load the dataset from a CSV file into a DataFrame
                                        # 'df' now contains all the data (likely includes bulb lifespans)
```

```python
from statsmodels.stats.weightstats import ztest  # Import the z-test function from statemodels
                                                 # This function will perform a one-sample z-test
```

```python
# Given data
population_mean = 1000  # Hypothesized population mean (H0): u = 1000)
                        # This is the value we are testing against
```

```python
alpha = 0.05  # Significance level (5%)
              # Therehold to decide whether to reject H0
              # If p-value < 0.05 -> reject null hypothesis
```

```python
#Step 1: Perform a One-sample Z-test
```

```python
z_stat,p_value = ztest(df["Lifespan (Hours)"],value = population_mean, alternative = "two-sided")
                    # ztest(...)perform a one-sample Z-test
```

```python
#Step 2: Decision
```

```python
if p_value < alpha:
    conclusion =("Reject the null hypothesis:The average lifespan is significantly different from 1000 hours")
else:
    conclusion =("Fail to reject the null hypothesis")
```

```python
print(f"Z-Statistics:{z_stat:.2f}")
print(f"P-Value:{p_value:.4f}")
print(f"Conclusion:{conclusion}")
```

```python
import pandas as pd
```

```python
from statsmodels.stats.weightstats import ztest
```

```python
# Load the datatest
df = pd.read_csv("ecommerce_two_sample_ztest_dataset.csv")
```

```python
print(df)
```

```python
# Separate the data into two groups
email_group = df[df['Group']=='Email Marketing']['Purchase_Value']
social_group = df[df['Group']=='Social Media Ads']['Purchase_Value']
```

```python
print ("Email Group: ")
print (email_group.describe())

print()
print("Social Group: ")
print(social_group.describe())
```

```python
# Perform the Two-Sample Z-test

z_test,p_value = ztest(email_group,social_group,alternative = 'two-sided')
```

```python
# Define the level of significance

alpha = 0.05
```

```python
# Decision based on P-value
if p_value < alpha:
    conclusion ="Reject the null hypothesis: There is a significance different in any purchase value"
else:
    conclusion ="Fail to reject the null hypothesis: There is no significance different in any purchase value"
```

```python
# Output Results
print(f"z-statistics:{z_stat:.2f}")
print(f"p-value:{p_value:.4f}")
print(f"Conclusion:{conclusion}")
```

```python
# TWO SAMPLE T-TEST
```

```python
from scipy import stats  #Import statistical functions(t-test)
```

```python
# Perform the two sample t-test(independent samples)

t_stat,p_value = stats.ttest_ind(email_group,social_group)
```

```python
#Significance Level

alpha = 0.05
```

```python
# Decision

if p_value < alpha:
    conclusion ="Reject the Null Hypothesis: There is a significant difference in average purchase value"
else:
    conclusion ="Fail to reject the Null Hypothesis: There is no significance difference in average purchase value"
```

```python
# Output

print(f"T-Statistics:{t_stat:.2f}")
print(f"P-Value:{p_value:.4f}")
print(f"Conclusion:{conclusion}")
```
