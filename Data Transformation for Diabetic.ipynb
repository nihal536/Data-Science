```python
import pandas as pd

df= pd.read_csv("Diab.csv")
print(df)
```

```python
df.info()
```

```python
df_cat= df[["Gender",'Diabetic']]
print(df_cat)
```

```python
df_cont = df.iloc[:,2:6]
print(df_cont)
```

```python
#STANDARDIZATION
# Standardization --> Converting data to have mean = 0 and standard deviation = 1

from sklearn.preprocessing import StandardScaler

SS = StandardScaler()

SS_X = SS.fit_transform(df_cont)

SS_X = pd.DataFrame(SS_X)

SS_X.columns =["OGTT","DBP","BMI","AGE"]

SS_X
```

```python
#SCALING/ NORMALIZATION
#Scaling/normalization --> Converting data into a fixed range (default: 0 to 1)

from sklearn.preprocessing import MinMaxScaler

MM = MinMaxScaler()

MM_X = MM.fit_transform(df_cont)

MM_X = pd.DataFrame(MM_X)

MM_X.columns = ["OGTT","DBP","BMI","AGE"]

MM_X
```

```python
#LABEL ENCODER
#Label Encoder converts categorical text data into numeric labels so machine learning models can process it

from sklearn.preprocessing import LabelEncoder

df_cat = df[["Gender","Diabetic"]].copy()

LE = LabelEncoder()

df_cat["Gender_LE"]=LE.fit_transform(df_cat["Gender"])

df_cat["Diabetic_LE"]=LE.fit_transform(df_cat["Diabetic"])

df_cat.head()
```

```python
#ONE HOT ENCODING
#One-Hot Encoding converting categorical values into multiple binary columns (0 and 1)
```

```python
from sklearn.preprocessing import OneHotEncoder

OHE = OneHotEncoder()

df_g = pd.DataFrame(OHE.fit_transform(df_cat[["Gender"]]).toarray())

df_g.columns = ['Female','Male']

print(df_g)
```

```python
df_d = pd.DataFrame(OHE.fit_transform(df_cat[["Diabetic"]]).toarray())

df_d.columns = ['NO','YES']

df_d
```

```python
# Combine Multiple DataFrames

X = pd.concat([SS_X,df_g,df_d],axis = 1)

X
```

```python
# Ordinal Encoding

import pandas as pd

from sklearn.preprocessing import OrdinalEncoder

# Load Dataset

df = pd.read_csv("Diab.csv")
df
```

```python
# Select Categorical Columns (example)

cat_cols = ['Gender','Diabetic']
```

```python
# Initialize Ordinal Encoder

oe = OrdinalEncoder()
```
