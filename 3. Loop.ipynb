```python
def display_customer_info(names,mobile_number):
    """Displays customer names and their corresponding mobile numbers
    Args:
        names: A list of customer names.
        mobile_numbers: A list ofcustomer mobile numbers"""
    if len(names) != len(mobile_number):
        print("Error : The number of names and mobile numbers do not match")

        return
    for names,mobile in zip(names,mobile_number):
        print("Customer : (names), Mobile : (mobile)")
```

```python
# Example Usage---
```

```python
customer_names =["Alice","Bob","Charlie"]
customer_mobiles =["123-456-7890","987-654-3218","555-123-4567"]

display_customer_info(customer_names,customer_mobiles)
```
