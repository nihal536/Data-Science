```python
#Define Feature Set and Target

x= df[['Daily Time Spent on Site','Age','Area Income','Daily Internet Usage','Male','Hour','Day']]
y= df['Clicked on Ad']
```

```python
from sklearn.linear_model import LogisticRegression

#Initialize and fit Logistic Regression model

logmodel = LogisticRegression(max_iter = 10000)
logmodel.fit(x,y)

#Predictions

predictions =logmodel.predict(x)
```

```python
#Evaluation

from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

conf_matrix = confusion_matrix(y,predictions)
accuracy = accuracy_score(y,predictions)

conf_matrix
```

```python
print("Accuracy score: ",accuracy)
```

```python
class_report = classification_report(y,predictions)
print(class_report)
```

```python
#Generate the accuracy score,recall,precision,f1 score,specificity

import numpy as np

from sklearn.metrics import recall_score,precision_score,f1_score
```

```python
#Assuming 'y' and 'predictions' are already defined from the previous code

#Calculate recall

recall =recall_score(y,predictions)
print("Recall: ",np.round(recall,2))
```

```python
#Calculate Precision

precision= precision_score(y,predictions)
print ("Precision: ",np.round(precision,2))
```

```python
#Calculate F1-Score

f1 = f1_score(y,predictions)
print ("F1-Score: ",np.round(f1,2))
```

```python
#Calculate Specificity

tn,fp,fn,tp =confusion_matrix(y,predictions).ravel()
# ravel() converts it into a flat list for easy unpacking

specificity = tn/(tn+fp)

print("Specificity: ",np.round(specificity,2))
```

```python
from sklearn.metrics import roc_curve,auc

predictions= logmodel.predict(x)
y_pred_proba= logmodel.predict_proba(x)[:,1]

fpr,tpr,thresholds= roc_curve(y,y_pred_proba)

roc_auc = auc(fpr,tpr)

plt.figure()
plt.plot(fpr,tpr,color='darkorange',lw=2,label=f'ROC curve(area=%0.2f)'%roc_auc)
```

```python
import numpy as np
import pandas as pd

from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
```

```python
#Load Dataset

df= pd.read_csv("Advertising1.csv")
```

```python
#Features (IndependentVariable) and target(DependentVariable)

# x --> input features used to predict.

x = df[['Daily Time Spent on Site','Age','Area Income','Daily Internet Usage']]

#y --> output(target variable)

y = df['Clicked on Ad']
```

```python
# Scale data (Importance for ML models)

scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)
```

```python
#Train Logistic Regression model

model = LogisticRegression(max_iter=10000)
model.fit(x_scaled,y)
```

```python
# STEP 1: Get model parameters

weights = model.coef_[0]
bias = model.intercept_[0]
```

```python
# STEP 2: Calculate Linear equation ==wx+b

z = np.dot(x_scaled,weights)+bias
```

```python
#STEP 3: Apply Sigmoid Function

def sigmoid(z):
    return (1/(1+np.exp(-z)))

probabilities = sigmoid(z)
```

```python
#STEP 4: Convert Probabilities into class labels

predictions =(probabilities >= 0.5).astype(int)
```

```python
#SHOW RESULTS

print ("First 10 Probabilities: \n",probabilities[:10])
print ("\n First 10 Predictions: \n",predictions[:10])
print ("\n Actual Values: ", y.values[:10])
```

```python
#==============================================================================================================
#Plot Sigmoid Curve

z_values = np.linspace(-10,10,100)
sign_values = sigmoid(z_values)

plt.figure()
plt.plot(z_values,sign_values)
plt.title("Sigmoid Function (s-curve)")
plt.xlabel=('z')
plt.ylabel=("probability")
plt.show()
```
