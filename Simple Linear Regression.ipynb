```python
import pandas as pd
```

```python
df = pd.read_csv("experience_vs_salary.csv")
```

```python
df
```

```python
from sklearn.linear_model import LinearRegression  # Import LinearRegression model from scikit-learn
                                                   # This model is used to predict continuous values (like salary, price, marks)

from sklearn.metrics import mean_squared_error, r2_score

                    # Import evaluation metrics:
                    # mean_squared_error -> measures average squared difference between actual and predicted value
                    # r2_score -> tells how well the model fits the data (goodness of fit)
```

```python
# Defining the independent (X) and dependent (y) variables

X= df[["Experience (Years)"]]   # Independent variable (input features)
                                # This is the column used to predict salary
                                # It is written in double brackets because sklearn expects a DataFrame (2D structure)

y= df["Salary ($)"]             # Dependent variable (target/output)
                                # This is what we want to predict
                                # Salary depends on experience, so it is the output variable

print(X)

print(y)
```

```python
# Creating and fitting the Linear Regression Model

model = LinearRegression()    # Create an instance (object) of the LinearRegression model
                              # This is the model that will learn the relationship between X and y

z = model.fit(X,y)            # Train the model using the data
                              # X = input (Experience)
```

```python
# Making predictions using the same dataset (without data partioning).

y_pred = model.predict(X)            # Use the trained model to predict output values (Salary)
                                     # X = input data (Experience)
                                     # y_pred = predicted salaries generated by the model
                                     # These are the model's estimated values based on the learned Equation

y_pred
```

```python
# Calculating R-squared (R2) and Mean Squaerd Error (MSE)

r2 = r2_score(y, y_pred)     # Calculate R-squared score
                             # Compares actual values (y) with predicted values (y_pred)
                             # Shows how well the model fits the data
                             # Value close to 1 -> good model

mse = mean_squared_error(y, y_pred)    # Calculate Mean Squared Error (MSE)
                                       # Measures average squared difference between actual and predicted values
                                       # Lower value -> better model performance
```
